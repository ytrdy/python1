```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.svm import SVC
```

```python
clf=SVC(gamma="auto")
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
Y=y
```

```python
X_1_2=x[:,:2]                  ###根據萼片長度&萼片寬度分析
X_3_4=x[:,2:4]                 ###根據花瓣長度&花瓣寬度分析
X_1_3=x[:,0:3:2]               ###根據萼片長度&花瓣長度分析
X_2_4=x[:,1:4:2]               ###根據萼片寬度&花瓣寬度分析
X_1_4=x[:,0:4:3]               ###根據萼片長度&花瓣寬度分析
X_2_3=x[:,1:3]                 ###根據萼片寬度&花瓣長度分析
```

```python
from sklearn.model_selection import train_test_split
```

### 萼片長度&萼片寬度

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(X_1_2,Y,
                                              test_size=0.2,
                                              random_state=0)
```

```python
plt.scatter(X_1_2[:,0],X_1_2[:,1],c=Y,cmap="Paired")
```

```python
clf.fit(x_train1,y_train1)
```

```python
y_predict1=clf.predict(x_test1)
```

```python
Y1=y_predict1-y_test1
```

```python
plt.scatter(x_test1[:,0],x_test1[:,1],c=Y1)
```

```python
x1=np.arange(3.8,8.2,0.02)
y1=np.arange(1.8,4.7,0.02)

xm1,ym1=np.meshgrid(x1,y1)
P1=np.c_[xm1.ravel(),ym1.ravel()]
z1=clf.predict(P1)

Z1=z1.reshape(xm1.shape)
plt.contourf(xm1,ym1,Z1,cmap="Paired",alpha=0.15)

plt.scatter(x_test1[:,0],x_test1[:,1],c=y_test1,cmap="Paired")
```
